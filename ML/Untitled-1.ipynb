```python
import pandas as pd

# Sample Route and IncidentMaster data
route_data = {
    'Route_Id': ['R0001', 'R0002'],
    'Total_Incidents_Count': [7, 8],
    'High': [1, 2],
    'Last_5Months_High': [0, 1],
    'Last_10Days_High': [1, 0],
    'Medium': [2, 1],
    'Last_5Months_Medium': [0, 0],
    'Last_10Days_Medium': [0, 1],
    'Low': [3, 1],
    'Last_5Months_Low': [1, 0],
    'Last_10Days_Low': [0, 1]
}

incident_master_data = {
    'Incident_Type': [
        'Health Emergencies', 'Fuel Costs', 'Storage Limitations', 'Overweight Cargo', 'Port Congestion', 
        'Shipping Lane Crowding', 'Communication Breakdown', 'Financial Instability', 'Lack of Standardization',
        'Documentation Errors', 'Limited Port Capacity', 'Container Mismanagement', 'Cargo Damage', 'Labor Strikes',
        'Regulatory Changes', 'Navigational Errors', 'Port Authority Regulations', 'Insurance Premiums',
        'Cargo Restrictions', 'Border Control Issues', 'Accidents at Sea', 'Port Accessibility', 'Economic Sanctions',
        'Quarantine Regulations', 'Sanitary Regulations', 'Legal Disputes', 'Waste Management', 'Supply Chain Disruptions',
        'Intermodal Transfer Delays', 'Freight Theft', 'Bunkering Delays', 'Insurance Issues', 'Trade Imbalances',
        'Equipment Failure', 'Container Shortages', 'Loading/Unloading Delays', 'Cargo Overbooking', 'Weather Conditions',
        'Political Instability', 'Technology Failures', 'Scheduling Conflicts', 'Documentation Fraud', 'Hull Fouling',
        'Customs Delays', 'Route Blockages', 'Insufficient Training', 'Cybersecurity Threats', 'Currency Fluctuations',
        'Piracy and Security', 'Environmental Regulations'
    ],
    'Severity': [
        'High', 'Medium', 'Low'
    ],
    'Description': [
        'Medical issues among crew affecting operations.', 'Fluctuations in fuel prices affecting operating costs.', 
        'Limited space for storing goods at ports.', 'Issues with cargo exceeding weight limits.', 
        'Delays caused by too many ships at a port.', 'High traffic in key shipping lanes causing delays.', 
        'Miscommunication between ship, port, and logistics providers.', 'Economic downturns affecting shipping demand.', 
        'Variations in port procedures and standards.', 'Incorrect or missing paperwork.', 
        'Insufficient infrastructure at ports.', 'Misplacement or mishandling of containers.', 'Damage to goods during transit.', 
        'Dockworker or crew strikes halting operations.', 'Sudden changes in shipping regulations.', 
        'Mistakes in navigation causing delays or accidents.', 'Compliance with various port-specific rules and regulations.', 
        'High insurance costs for high-risk areas or cargo.', 'Limitations on certain types of cargo.', 
        'Delays at international borders.', 'Collisions, groundings, or other maritime accidents.', 
        'Challenges with accessing certain ports due to size or draft restrictions.', 
        'Trade restrictions affecting shipping routes and schedules.', 'Delays due to health and safety inspections.', 
        'Compliance with health and hygiene standards for certain cargo types.', 
        'Litigation over contracts, cargo, or accidents.', 'Managing waste and pollution from ships.', 
        'Interruptions in the supply chain affecting schedules.', 
        'Delays in transferring goods between different transport modes.', 'Theft of cargo during transit.', 
        'Delays in refueling ships.', 'Problems with cargo insurance claims or coverage.', 
        'Discrepancies in import/export volumes affecting schedules.', 'Malfunctions in ships, cranes, or other equipment.', 
        'Lack of available containers for shipping.', 'Inefficiencies in cargo handling processes.', 
        'Overbooking of cargo space leading to delays.', 'Adverse weather like storms or fog affecting schedules.', 
        'Political events affecting shipping routes and schedules.', 'Failures in logistics management software or tracking systems.', 
        'Overlapping schedules for ships or port facilities.', 'Fraudulent documents or illegal shipments.', 
        'Growth of organisms on hulls reducing ship efficiency.', 'Slow processing of goods through customs.', 
        'Blocked shipping routes due to accidents or geopolitical issues.', 'Lack of proper training for crew or port workers.', 
        'Risks of cyber-attacks on shipping systems.', 'Exchange rate changes affecting costs.', 
        'Threats from pirates or terrorist activities.', 'Compliance with emission standards and pollution controls.'
    ]
}

routes_df = pd.DataFrame(route_data)
incident_master_df = pd.DataFrame(incident_master_data)

# Generate Incident Excel data
incident_data = []
incident_id = 1

for _, route in routes_df.iterrows():
    route_id = route['Route_Id']
    incidents_count = route['Total_Incidents_Count']
    high = route['High']
    last_5months_high = route['Last_5Months_High']
    last_10days_high = route['Last_10Days_High']
    medium = route['Medium']
    last_5months_medium = route['Last_5Months_Medium']
    last_10days_medium = route['Last_10Days_Medium']
    low = route['Low']
    last_5months_low = route['Last_5Months_Low']
    last_10days_low = route['Last_10Days_Low']
    
    for _, incident in incident_master_df.iterrows():
        if incidents_count == 0:
            break
        
        if high > 0 and incident['Severity'] == 'High':
            severity_label = 'High'
            if last_5months_high > 0:
                severity_label = 'Last_5Months_High'
                last_5months_high -= 1
            elif last_10days_high > 0:
                severity_label = 'Last_10Days_High'
                last_10days_high -= 1
            high -= 1
        elif medium > 0 and incident['Severity'] == 'Medium':
            severity_label = 'Medium'
            if last_5months_medium > 0:
                severity_label = 'Last_5Months_Medium'
                last_5months_medium -= 1
            elif last_10days_medium > 0:
                severity_label = 'Last_10Days_Medium'
                last_10days_medium -= 1
            medium -= 1
        elif low > 0 and incident['Severity'] == 'Low':
            severity_label = 'Low'
            if last_5months_low > 0:
                severity_label = 'Last_5Months_Low'
                last_5months_low -= 1
            elif last_10days_low > 0:
                severity_label = 'Last_10Days_Low'
                last_10days_low -= 1
            low -= 1
        else:
            continue
        
        incident_data.append({
            'Incident_Id': f'INC{incident_id:03d}',
            'Route_Id': route_id,
            'Incident_Type': incident['Incident_Type'],
            'Severity': severity_label,
            'Description': incident['Description']
        })
        incident_id += 1
        incidents_count -= 1

# Convert the data into a DataFrame and save as an Excel file
incident_df = pd.DataFrame(incident_data)
incident_df.to_excel('Incident.xlsx', index=False)

print("Incident Excel sheet generated successfully.")
```
