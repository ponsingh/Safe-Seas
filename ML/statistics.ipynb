```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
from datetime import datetime, timedelta

def get_random_date_in_last_two_years():
    """Generates a random date within the last two years."""
    today = datetime.now()
    two_years_ago = today - timedelta(days=2 * 365)  # Roughly account for leap years

    # Get random time difference within the two-year range
    random_days = random.randrange((today - two_years_ago).days)

    # Add the random time difference to the start date
    return two_years_ago + timedelta(days=random_days)
```

```python



routes = ["R" + str(i).zfill(3) for i in range(1, 701)] 


# R001 to R700
incident_types = ["Piracy", "Theft", "Accident", "Natural Disaster", "Delay","Attacks","Port Security"]
severities = ["High", "Medium", "Low"]
theftMessage = ["Cargo theft","Container theft"]
accMessgae=["Minor collision"
,"Offshore Oil Rig Mishaps						"
,"Accidents on Tugboats                          "
,"Accidents on Crude Oil Tankers and Cargo Ships "
,"Maritime Accidents on Diving Support Vessels:  "
,"Accidents on Barges                            "
,"Minor collision                                "
,"Offshore Oil Rig Mishaps                       "
,"Grounding of Ships                             "
,"Accidents on Crude Oil Tankers and Cargo Ships "
,"Maritime Accidents on Diving Support Vessels:  "
,"Accidents on Barges                            "
,"Offshore Oil Rig Mishaps                       "
,"Accidents on Tugboats                          "
,"Maritime Accidents on Diving Support Vessels:  "
,"Accidents on Barges                            "]

data = []
for route in routes:
    for _ in range(random.randint(2, 10)):  # 2-4 incidents per route
        incident_type = random.choice(incident_types)
        severity = random.choice(severities)
        incident_date=get_random_date_in_last_two_years()
        description=""
        # ... Generate other data (location, date, etc.)
        if(incident_type=="Theft"):
            description=random.choice(theftMessage)
        elif incident_type=="Accident":
            description=random.choice(accMessgae)
            



        data.append([route, incident_type, severity,'',incident_date,description])

df = pd.DataFrame(data,
                  columns=["RouteID", "IncidentType", "Severity","Location","IncidentDate","Description"])

new_series = pd.Series(["INC" + str(i).zfill(3) for i in range(1, len(df))] )
df["Incidents"]= new_series
df.to_clipboard()
```
